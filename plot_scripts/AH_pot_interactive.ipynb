# Temperature dependence of Ashbaugh-Hatch Potential

Extension of the Ashbaugh-Hatch Potential by implementing the temperature dependence to cover interactions between hydrophobic residues in proteins based on the derivation shown in [_H. Wennerstrom & B. Lindman_](https://doi.org/10.1016/j.molliq.2025.128169). 

_S. Pineda Pineda & M. Lund, 2025_.

---

### Generalized Ashbaugh–Hatch Potential

The potential is defined as:

$$
\begin{aligned}
U(r) &= 
\begin{cases} 
U_{\mathrm{LJ}}(r) + \epsilon(T) (1 - \lambda) - \lambda\, U_{\mathrm{LJ}}(r_{\mathrm{cutoff}}), 
& r \le 2^{1/6}\sigma \\[2mm]
\lambda \left[ U_{\mathrm{LJ}}(r) - U_{\mathrm{LJ}}(r_{\mathrm{cutoff}}) \right], 
& r > 2^{1/6}\sigma \\[1mm]
0 & r \ge r_{\mathrm{cutoff}}
\end{cases} \\[2mm]
\end{aligned}
$$

where:

- $U_{\mathrm{LJ}}(r)$ is the standard Lennard–Jones (LJ) potential.
- $\epsilon(T)$ corresponds to the depth of the LJ potential well, which in our case is a function of T (see below).
- $\lambda$ is a dimensionless parameter that controls the interpolation between the standard LJ potential ($\lambda=1.0$) and the Week-Chandler-Andersen potential ($\lambda=0.0$).
- $\sigma$ represents the effective size of the particle. In the interactive example shown below, it is set to 1.


### Temperature-dependent energy well

Using as a template Eq.15 described in [_H. Wennerstrom & B. Lindman_](https://doi.org/10.1016/j.molliq.2025.128169).

$$
\begin{aligned}
\epsilon(T) &= \epsilon_c \Bigg[
T \left( \frac{1}{T_c} + c - c \ln \frac{T}{T_c} \right) - c\,T_c
\Bigg]
\end{aligned}
$$

The reference value for the minimum depth of the potential, $\epsilon_c = 0.8368$ at $T_c = 293.0$  was taken from [_G. Tesei & K. Lindorff-Larsen_](https://doi.org/10.12688/openreseurope.14967.2). For alkanes, $c=\frac{\Delta C_p}{\Delta G} \approx 1.9\cdot 10^{-2}\, K^{-1}$ (Eq. 12 in [_H. Wennerstrom & B. Lindman_](https://doi.org/10.1016/j.molliq.2025.128169)), for amino acid side chains, $c\approx 2.52\cdot 10^{-2}\, K^{-1}$ based on the thermodynamic properties reported in [_G. I. Makhatadze_](https://doi.org/10.1016/S0301-4622(98)00095-7) and [_V. Pliška, et al._](https://doi.org/10.1016/S0021-9673(00)82337-7).
